```python
fig , ax = plt.subplots()
movie_type=["action", "comedy", "thriller", "scifi"]
men_reponses=[3,5,9,12]
women_response=[7,3,6,10]
ax.bar(movie_type, men_responses, color='b', width=0.25)
ax.bar(movie_type, women_response, bottom= men_responses, color='r', width = 0.25)
ax.set_ylabel("Survey response")
ax.set_xticklabels(movie_type, rotation=90)
ax.legend(labels=["men","women"])
plt.show()
```
